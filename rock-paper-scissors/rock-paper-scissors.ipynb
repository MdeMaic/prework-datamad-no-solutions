```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list
import random
import math

# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.
game_list = ['stone', 'paper', 'scissors']

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...
max_games = [1,3,5,7,9]
n_partidas = int(input("Elige el número de partidas: "))
while n_partidas not in max_games:
    print ("Debes elegir entre",str(max_games),"partidas")
    n_partidas = int(input("Elige el número de partidas: "))
    
# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games
to_win = (n_partidas/2)+ 0.1

# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.
def machine_game():
    return(random.choice(game_list))

# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.
def your_choice():
    answer = input("La máquina ya ha elegido - Cuál es tu elección: ")
    while answer not in game_list:
        print("Tu elección no es correcta, elige entre "+str(game_list))
        answer = input ("Cuál es tu elección: ")
    return(answer)

# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins
def battle_result(machine,player):
    result = 0
    if machine == player:
        result = 0
    elif (machine == "stone" and player == "scissors") or (machine == "scissors" and player == "paper") or (machine == "paper" and player == "stone"):
        result = 1
    elif (player == "stone" and machine == "scissors") or (player == "scissors" and machine == "paper") or (player == "paper" and machine == "stone"):
        result = 2
    return(result)

# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated
def state_of_game(machine,player,this_game_result,machine_wins,player_wins):
    print("PC:",machine,"   -   PLAYER:",player)
    if this_game_result == 0:
        print("Hay un empate en esta partida")
    elif this_game_result == 1:
        print("Gana la máquina, sorry.")
        machine_wins += 1
    elif this_game_result == 2:
        print("Has ganado un punto!")
        player_wins += 1
    print("PC - PLAYER")
    print(machine_wins,"    ",player_wins)
    
# Create two variables that accumulate the wins of each participant
player_win = 0
pc_win = 0

# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.
print("\nHas elegido",n_partidas,"partidas - Gana el que alcance primero los",math.ceil(to_win),"puntos o el mejor de",n_partidas)
print("QUE COMIENCE EL JUEGO")
for i in range(n_partidas):
    print("\nPARTIDA",i+1)
    pc = machine_game()
    play = your_choice()
    res = battle_result(pc,play)
    state_of_game(pc,play,res,pc_win,player_win)
    if res == 1:
        pc_win += 1
    elif res == 2:
        player_win +=1
    
    if pc_win > to_win or player_win > to_win:
        print("FIN DE LA PARTIDA")
        break

# Print by console the winner of the game based on who has more accumulated wins
if pc_win > player_win:
    print("WASTED - Has perdido")
elif pc_win < player_win:
    print("ENHORABUENA! - Has ganado")
elif pc_win == player_win:
    print("El juego ha terminado en empate!")
```
